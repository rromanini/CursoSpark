```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
#Estilos disponiveis
print(plt.style.available)
```

```python
def cria_plot():
    x = np.random.randn(5000,6)
    (figure,axes) = plt.subplots(figsize=(16,10))
    (n,bins,patches) = axes.hist(x,12,normed=1,histtype="bar",label=["Color1","Color2","Color3","Color4","Color5","Color6"])
    axes.set_xlabel("Dados",fontsize=16)
    axes.set_ylabel("Frequencia",fontsize=16)
    axes.set_title("Histograma para distribuição normal",fontsize=25)
    axes.facecolor="2b3e50"
    axes.labelcolor="ff812d"
    

    axes.legend()
    plt.show()
```
